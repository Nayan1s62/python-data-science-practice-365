# 16. (Hard) Take a sentence. Split into words, remove all words that start with vowel
# (a,e,i,o,u), then join remaining words.

```python

line = input("enter")
word = "".join([word for word in line.split() if word[0].lower()not in "a,e,i,o,u"])
print(word)
```

```python
# 19. (Hard) Take a paragraph. Split it into sentences (using .) strip each sentence,
# and join them with "\n- ".
```

```python
graph = ("I am the master mind.  who deciding at a time. what to do when to do in my life ")
graph.split(".")
```

```python
graph.strip()
```

```python
graph = "I am the master mind. who deciding at a time. what to do when to do in my life "

graph = graph.strip()
graph = graph.split(".")

new = []

for sentence in graph:
    new.append(sentence.strip())

result = "\n- ".join(new)

print(result)
```

# 20. (Hard) Take a string. Use replace(), strip(), split(), and join() together to clean
# and reformat it as a proper sentence.

```python
text = input("Enter a sentence: ")

text = text.replace("_", " ")
text = text.strip()
words = text.split()
text = " ".join(words)

print(text)
```
