# sets {} operation that important for it 
1) union     # all unique elements from both sets
2) Intersection # common elements from both sets 
3) Difference  # elements in A but not in B 
4) Symmetric   # elements not common 
5) Difference  # elements in B but not in A 


# union (X) : 

```python
A = {1,2,3,4}
B = {3,4,5,6}

C = A.union(B)
print(C)
```

```python
s2 = {"nayan","yash","ramu", "shamu"}
s3 = {"punit","himanshu","ramu","shamu"}
s4 = s2.union(s3)
s4
```

# Intersection (X) : - common elements from both sets 

```python
A = {1,2,3,4}
B = {3,4,5,6}

C = A.intersection(B)       # print common elements from both 
C
```

```python
s2 = {"nayan","yash","ramu", "shamu"}
s3 = {"punit","himanshu","ramu","shamu"}
S4 = s2.intersection(s3)
S4
```

```python
s2 = {"nayan","yash","ramu", "shamu"}
s3 = {"punit","himanshu","ramu","shamu"}
s4 = s2.difference(s3)        # aise elements jo S2 me hai but s3 me nai hai vo print honge 
s4
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

c = A.difference(B)
print(c)
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

D = B.difference(A)   # aise elements jo b me hai but a me nai hai vo print honge 
D
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

c = A.symmetric_difference(B)   # aise elements jo dono set me common nai hai aise elements display honge jo ek hi baar aye hai set me
c
```

# sets {} operation that important for it  we can use this also 
1) union   # ( A|B)   # all unique elements from both sets
2) Intersection # A & B ) common elements from both sets 
3) Difference  #   A - B ) elements in A but not in B 
4) Symmetric   #  A ^ B ) elements not common 
5) Difference  #   elements in B but not in A
