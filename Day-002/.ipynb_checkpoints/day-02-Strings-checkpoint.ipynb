# 13. Take a password as input and check:
# • It should be alphanumeric
# • It should be minimum 8 characters long Print "Valid Password" or
# "Invalid Password".

```python
password = (input("enter password"))
if len(password) >= 8 and password.isalnum():  # len(password) → Counts the number of characters in the password.
    print("valid password")                  # Checks whether the password has at least 8 characters.
else :                                        #   and → Both conditions must be True for the password to be considered valid.
    print("Invalid password")
```

# 20. Take any string from user.
# • Print the string in reverse order using slicing
# • Convert it to title case
# Check if the reversed string is same as original (Palindrome check)

```python
#None ka matlab hai:

#"Default value use karo."

#Agar step positive hai (1) to:
#start = 0
#stop = len(string)
#Agar step negative hai (-1) to:
#start = last index
#stop = beginning se ek pehle (internally -len(string)-1 jaisa behavior)

user = input("enter :- ")
n = slice (None,None,-1)   
print(user [n] )
```

```python

user = input("enter :- ") 
n = slice (-1,-5)     # Python cannot move forward from index -1 to -5, so it returns nothing.
print(user[n]) 
```
