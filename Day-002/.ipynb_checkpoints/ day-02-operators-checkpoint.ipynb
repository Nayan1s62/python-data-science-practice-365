```python
# Question 1: Username Checker (String + if-else)

#  Take a username from the user.

# Rules:

# If the username is exactly "nayan", print "Welcome Nayan!"
# Otherwise, print "Access Denied"
name = input("enter user name ")
if name == "nayan ":
    print("welcom nayan bro ")
else :
    print("Access Denied ")
```

```python
# Question 2: Password Length Checker (String + if)

# Take a password from the user.

# If the password has 8 or more characters, print:

password = input("Enter password: ")

if len(password) >= 8:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
# 3. First Letter Vowel or Consonant
word = input("Enter a word: ")

first_letter = word[0].lower()

if first_letter in "aeiou":
    print("Starts with a vowel")
else:
    print("Starts with a consonant")
    
```

```python
# 4. Name Starts with 'A'

name = input("Enter your name: ")

if name[0].lower() == "a":
    print("Your name starts with A")
else:
    print("Your name does not start with A")
```

```python
# 5. Login System
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "python123":
    print("Login Successful")
else:
    print("Invalid Username or Password")
```

```python
# ⭐ Bonus Challenge
sentence = input("Enter a sentence: ")

if "python" in sentence.lower():
    print("Python found!")
else:
    print("Python not found!")
```
