# 1. Take a number as input. If the number is positive, check whether it is even or odd
# and print the result.

```python
A = int(input("enter a number "))

if A % 2 == 0 :
        print("The number is even :-",A)
else:
        print("The number is ODD: - ",A)
```

# 2. Take age as input. If age is greater than or equal to 18, check if the person is a
# senior citizen (age >= 60) or adult.

```python
age = int(input("enter the user age "))
if age >= 18 :
    print("The adult ")
if age >= 60 :
    print("the senior citizen")
```
