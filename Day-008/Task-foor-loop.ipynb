# 5. Print the following pattern using for loop:

```python

for i in range (1,  10):
  for j in range(i):
     print("*",end= "")
     print()   
```

```python
for i in range(1, 10):
    for j in range(i):
        print("*", end="")
    print()
```

# Q8)Count numbers divisible by 3 between 1 and 50

```python
count = 0

for i in range(1, 51):

    if i % 3 == 0:
        count = count + 1

print("Total =", count)
```
