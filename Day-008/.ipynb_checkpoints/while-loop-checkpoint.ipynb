# Write a while loop to print numbers from 1 to 15.

# Write a while loop to print numbers from 1 to 15.
# Logic

```python
num = 1
while num <= 15:
    print(num)
    num += 1
```

# Question 12
# Multiplication Table using while loop

```python
number = int(input("Enter number: "))

i = 1

while i <= 10:
    print(number, "x", i, "=", number * i)
    i = i + 1
```

# Question 13
# Sum numbers until user enters 0

```python
### total = 0

number = int(input("Enter number (0 to stop): "))

while number != 0:
    total = total + number
    number = int(input("Enter number (0 to stop): "))

print("Total =", total)
```

# Question 14
# Print odd numbers from 1 to 20
# Logic

```python
i = 1

while i <= 20:
    print(i)
    i = i + 2
```

# Question 15
# Divide by 2 until number becomes less than 1
# Logic

```python
number = float(input("Enter number: "))

count = 0

while number >= 1:

    number = number / 2

    count = count + 1

print("Steps =", count)
```
