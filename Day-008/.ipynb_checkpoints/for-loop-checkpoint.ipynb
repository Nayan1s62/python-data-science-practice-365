```python
lists1 = ["sumit","sumit","sumit123","SuMiT","Sumit@123","1234"]

for word in lists1 :
    if word.isalpha():
        print("Alpha")
    elif word.isdigit():
        print("digit")
    elif  word.isalnum():
        print("alpha+num")
    elif  word.islower():
        print("lower")
    else:
        print("special word")
```
