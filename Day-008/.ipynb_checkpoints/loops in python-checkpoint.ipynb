```python
num = 1                # start kaha se kare 
while num <= 10:        # kaha stop kare loop 
    print(num)
    num += 1           # increment kaha krna hai last part 
```

```python
number = 0                       # take the no 0
even_odd = []                     # take empty list 
while number <= 10:                  
                                            # stop condition loop part unit 11
    if number % 2 == 0 :                   # block of code 
        even_odd.append("Even")
    else:
        even_odd.append("odd")
        number += 1

    print(even_odd)                     # always print outside of loop part list 
    
```

```python
num = 12345                  # start 12345 
rev = 0                      # 0 , 5 

while num > 0 :                 # it is 5    

    digit = num % 10              #  5 value # module return last digit  # this is important part 

    rev = rev * 10 + digit        # store 12345 01234 store 5  # 

    num = num // 10     # 12345 after // we get 1234      # he remove the last digit  # he removes the part 

print(rev)         
```

```python
original = 12345
num = 12345                  # start 12345 
rev = 0                      # 0 , 5 

while num > 0 :                 # it is 5    

    digit = num % 10              #  5 value # module return last digit  # this is important part 

    rev = rev * 10 + digit        # store 12345 01234 store 5  # 

    num = num // 10   
    if num == original :
        print ("palidrome") 

    else:
        print("it is not a palidrome ")

print(rev)         
```

```python
num = 12345

original = num 
rev = 0 

while num > 0 :

    digit = digit % 10 
rev = rev * 10 + digit 
num = num // 10 
if original == rev :
     result = ("palidrome")
else:
     result = ("not palidrome")

result
```
