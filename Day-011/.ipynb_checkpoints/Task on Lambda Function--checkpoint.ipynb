# Write a lambda function to return the square of a number.

```python
s = lambda  a : a **2 
s(5)
```

```python
# 2.
# Write a lambda function to return the cube of a number.
```

```python
cube = lambda  C : pow(C,3)
cube(4)
```

```python
# 3.
# Write a lambda function to add two numbers.
```

```python
add = lambda a,b: [a + b ]
print(add(20,26))
```

```python
# 4 
# Write a lambda function to find the maximum of two numbers.
```

```python
yash = lambda a,b,c: max(a,b,c)
print(yash(1,2,5))
```

```python
# Write a lambda function that returns "Even" if a number is even; otherwise "Odd".
```

```python
even = lambda x : "Even " if x % 2 == 0 else "odd"
even(5)
```

```python
# Use map() with lambda to square every element of a list.
```

```python
mapu = [1,2,3,4,5,6,7,8]
result= list(map(lambda i : i*2 ,mapu))
result
```

```python

# 7.
# Use filter() with lambda to keep only even numbers.
```

```python
mapu = [1,2,3,3,4,5,6,7,8]
result= list(filter(lambda i :  i % 2 == 0 ,mapu))
result
```

```python
# 8.
# Use filter() with lambda to keep numbers greater than 50.
 # Input:
#[20,45,55,70]

list1 = [20,45,55,70]
list(filter(lambda x : x >= 50 , list1))
```
