```python
# lambda inside map () function 

number = [1,2,3,4,5,6]

result = list(map(lambda x : "even"if x % 2 == 0 else "odd",number))     # mention the number  from where to select , # if not use map then () paranthesis is used 
result
```

# lambda inside the filter(): - function 

```python
number = [1,2,3,4,5,6]

result = list(filter(lambda x : "even"if x % 2 == 0 else "odd",number))     # mention the number  from where to select , # if not use map then () paranthesis is used 
result
```
