```python
# 7. Write a function manual_deep_copy(lst) that manually creates a deep copy of a nested list
# without using copy.deepcopy().
```

```python
import copy

def manual_deep_copy(lst):
    new_list = []

    for inner_list in lst:

        inner_copy = []

        for item in inner_list:
            inner_copy.append(item)

            new_list.append(inner_copy)
            
            return new_list

list1 = [[1,2],[3,4]]
list2 = manual_deep_copy(list1)

list2[0][0] = 100

print("Original List :",list1)
print("copied List :",list2)
```

```python
# 9. (Hard) Create a function that detects whether a copy is shallow or deep by modifying a
# nested element and checking the original.
```

```python
def check_copy(original,copied):

    copied[1][0]=100

    if original [1][0] == 100:
        print("it is shallo copy :-")
    else:
        print("it is deep copy")

data = [10, [20, 30], 40]

shallo = copy.copy(data)
deep = copy.deepcopy(data)
check_copy(data,shallo)

data = [10,[20,30],40]
check_copy(data,deep)
```

```python
# 10. (Hard) Create a class Student with attributes name and marks (list). Demonstrate what
# happens when you do shallow copy vs deep copy of a Student object.
```

```python
import copy

class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks


student1 = Student("Nayan", [80, 90, 70])

student2 = copy.copy(student1)

student3 = copy.deepcopy(student1)

student2.marks[0] = 100
student3.marks[1] = 50

print("Original :", student1.marks)
print("Shallow  :", student2.marks)
print("Deep     :", student3.marks)
```
