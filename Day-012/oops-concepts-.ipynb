```python
# without constructor 
# self is used to share attribute for many methods 
class Students:

# instance method 
    def student_info(self,name,roll_no,class_n):                 # create the argument 

       self.Name = name 
       self.Roll_number = roll_no
       self.Class_name = class_n
   # method 2:-
    def display_information(self):
        
       print(f"Student Name : {self.Name}")
       print(f"Student roll_no : {self.Roll_number}")
       print(f"Student class_n : {self.Class_name}")

# creating the clss object

std1 =Students() 

# call the method 1 : 
std1.student_info("Punit",420,"Naresh-IT")
```

```python
# with - constructor 
# self is used to share attribute for many methods 
class Students:

# instance method 
    def _init_(self,name,roll_no,class_n):                 # create the argument 

       self.Name = name 
       self.Roll_number = roll_no
       self.Class_name = class_n
   # method 2:-
    def display_information(self):
        
       print(f"Student Name : {self.Name}")
       print(f"Student roll_no : {self.Roll_number}")
       print(f"Student class_n : {self.Class_name}")

# creating the clss object # when 

std1 =Students("Nayan",456,"nareshit")   #in constructor

# call the method 1 : 
std1.display_information()
```
