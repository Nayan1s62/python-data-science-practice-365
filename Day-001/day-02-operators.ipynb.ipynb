# 12. Student Grade Calculator #

```python
maths = int(input("enter maths marks "))
english = int(input("enter the english marks "))
marathi = int(input("enter the marathi marks "))
science = int(input("enter the science marks "))
hindi   = int(input("enter the hindi marks "))
percentage = int(maths + english + marathi + science + hindi)/5
print(percentage)
if (percentage >=90):
    print("grade A+ ")

elif(percentage >=80):
    print("grade A ")
    
elif(percentage >=70):
    print("grade B")
    
elif(percentage >=60):
    print("grade C")
    
elif(percentage <=50):
    print("FAIL")
print (percentage)
```
